```python
from networkx.generators.community import LFR_benchmark_graph
import networkx as nx
import igraph as ig
import subprocess
```

```python
vertices=1000
```

```python
def netx_to_igraph(G):
    g = ig.Graph(n=G.number_of_nodes(), directed=False)
    g.add_edges(G.edges())
    g.vs["name"] = range(G.number_of_nodes())
    return g

def modularity(g):
    print(g.community_multilevel())
    return

def infomap(g):
    info=g.community_infomap()
    print(info)
    return

def run_maxperm(g):
    
    input_file = open("input.dat", "w")
    input_file.write(str(g.vcount()) + " " + str(g.ecount()) + "\n")
    for i in g.es:
        m1 = i.tuple[0]
        m2 = i.tuple[1]
        input_file.write(str(m1) + " " + str(m2) + "\n")
#         input_file.write(str(m2) + " " + str(m1) + "\n")
    input_file.close()
    
    returned_value = subprocess.check_output("bash compile.sh", shell = True)
    returned_value = returned_value.decode("utf-8")
    
    #print(returned_value)
    
    output_file = open("output.txt","r")
    partitions = output_file.readlines() 
    perm_arr=[]
    perm_community = []
    
    for line in partitions[2:]:
        #print(line.strip())
        perm_arr.append(int(line.split('\t')[-1].strip()))
    
    dicti={}
    for i in range(len(perm_arr)):
        comm = []
        j=i
        comm.append(j)
        while(not (perm_arr[j] == j)):
            j=perm_arr[j]
            comm.append(j)
        if comm[-1] in dicti:
            for ele in comm:
                dicti[comm[-1]].add(ele)
        else:
            dicti[comm[-1]]=set(comm)
    
    print(dicti)
    print(len(dicti))
```

```python
G = LFR_benchmark_graph(n=vertices,tau1=2,tau2=1.5, mu=0.4, average_degree=10, max_degree=50, min_community = None, max_community = None, tol = 1e-07, max_iters = 500, seed = None)
#nx.draw(G, with_labels=True)
```
