```python
from igraph import *
import networkx as nx
from random import randint
```

```python
def Print(g,num_nodes):
  G = nx.Graph()
  for i in range(num_nodes):
    G.add_node(i)
  for i in range(num_nodes):
    for j in range(i+1, num_nodes):
      if g[i,j] == 1:
        G.add_edge(i,j)
  nx.draw(G, with_labels=True)
```

```python
def infomap(g):
    k = g.community_infomap()
    print(k)
```

```python
def ring_cliques(l):
    num_nodes = 0
    for x in l:
        num_nodes += x
    g = Graph.Full(num_nodes)
    for i in range(num_nodes - 1):
        for j in range(i+1,num_nodes):
            g[i,j] = 0
    size = 0
    for x in l:
        for i in range(size, size+x-1):
            for j in range(i+1, size+x):
                g[i,j] = 1
        if(size>0):
            g[size,size-1] = 1
        size += x
    g[0, num_nodes-1] = 1
    Print(g,num_nodes)
    infomap(g)
```

```python
# Rerun this cell for multiple trials
l = []
for i in range(randint(4,10)):
    l.append(randint(5,25))
print(l, len(l))
ring_cliques(l)
```
